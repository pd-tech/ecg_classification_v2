```python
import sys
import csv
import matplotlib.pyplot as plt
from numpy import *
#import numpy as np

time = []
volts = []
marker = []

file = "ciste;+R-L;rec_U01_Wed_Oct_31_09_51_14_2018_.txt" #"../ekg/ciste;+R-L;rec_U01_Wed_Oct_31_09_51_14_2018_.txt"
time_step = 0.002000 # 500 Hz

with open(file, newline='') as csvfile:
    time.append(float(0.000000))
    spamreader = csv.reader(csvfile, delimiter=',')
    for row in spamreader:
        if time: time.append(time[-1]+time_step)
        volts.append(float(row[6]))
```

```python
start = 50000 # if len(sys.argv)<3 else int(sys.argv[2])
end = 55000 #len(time) # if len(sys.argv)<4 else int(sys.argv[3])

print(mean(volts[start:end]))
expected_value = mean(volts[start:end])

volts2 = []
for i in volts[start:end]:
    volts2.append(i - expected_value)
    #print(i)


fig, ax = plt.subplots()
#plt.autoscale(False)
ax.plot(time[start:end], volts2)#volts2[start:end])
#ax.plot(time[start:end], volts[start:end])



plt.show()
```
