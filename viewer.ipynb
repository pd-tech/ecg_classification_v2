```python
import sys
import csv
import matplotlib.pyplot as plt
from numpy import *
from itertools import islice
#import numpy as np

time = []
volts = []
marker = []

file = "ciste;+R-L;rec_U01_Wed_Oct_31_09_51_14_2018_.txt" #"../ekg/ciste;+R-L;rec_U01_Wed_Oct_31_09_51_14_2018_.txt"
time_step = 0.002000 # 500 Hz

with open(file, newline='') as csvfile:
    spamreader = islice(csv.reader(csvfile, delimiter=','), 7, None, None) # Skip the header of length of 7 lines
    for row in spamreader:
        if not time: time.append(float(0.000000)); volts.append(float(row[6]))
        else: time.append(time[-1]+time_step); volts.append(float(row[6]))
        #volts.append(float(row[6]))
```

```python
start = 100000 # if len(sys.argv)<3 else int(sys.argv[2])
end = 105000 #len(time) # if len(sys.argv)<4 else int(sys.argv[3])

expected_value = mean(volts[start:end])
print("Expected value=" + str(expected_value))

volts2 = []
for value in volts[start:end]:
    volts2.append(value - expected_value)
    #print(i)


fig, ax = plt.subplots()
#plt.autoscale(False)
ax.plot(time[start:end], volts2)#volts2[start:end])
#ax.plot(time[start:end], volts[start:end])

plt.show()
```
